## SOM

```python
def visualize_som(som, columns, labels = ["", ""], dim = 10):
    reshaped_x = som.w[:, columns[0]].reshape(dim, dim)
    reshaped_y = som.w[:, columns[1]].reshape(dim, dim)
    for v in range(dim):
        plt.plot(reshaped_x[:, v].numpy(), reshaped_y[:, v].numpy(), marker='None', linestyle='-', color='blue')
        plt.plot(reshaped_x[v, :].numpy(), reshaped_y[v, :].numpy(), marker='None', linestyle='-', color='blue')
    plt.plot(som.w[:, columns[0]].numpy(), som.w[:, columns[1]].numpy(), marker='o', color='blue')
    plt.xlabel(labels[0])
    plt.ylabel(labels[1])
    plt.show()
```

### State Weight Visualization

```python
state_filehandler = open("../data/pose_somatotopic/cartpole/ppo/state_som.obj", 'rb')
state_model = pickle.load(state_filehandler)
state_model.w.shape
```

```python
visualize_som(state_model, [0, 1], ["Cart Position", "Cart Velocity"])
```

```python
visualize_som(state_model, [1, 2], ["Cart Velocity", "Pole Angle"])
```

```python
visualize_som(state_model, [2, 3], ["Pole Angle", "Pole Angular Velocity"])
```
